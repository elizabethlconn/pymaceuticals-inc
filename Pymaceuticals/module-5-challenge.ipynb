# Prepare Data
1) Run the provided package dependency and data imports, and then merge the mouse_metadata and study_results DataFrames into a single DataFrame.
2) Display the number of unique mice IDs in the data, and then check for any mouse ID with duplicate time points. Display the data associated with that mouse ID, and then create a new DataFrame where this data is removed. Use this cleaned DataFrame for the remaining steps.
3) Display the updated number of unique mice IDs.

```python
# Dependencies and Data Imports:
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Data File Paths:
mouse_metadata_path = 'data/Mouse_metadata.csv'
study_results_path = 'data/Study_results.csv'

# Read Data:
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Merge DataFrames:
pymaceuticals_df = pd.merge(study_results, 
                            mouse_metadata, 
                            how = 'left', 
                            on = ['Mouse ID', 'Mouse ID']
                           )

# Number of Unique Mouse IDs:
unique_mice_id = pymaceuticals_df['Mouse ID'].nunique()
print(f'\n\033[4mNumber of unique Mouse IDs\033[0m: {unique_mice_id}')

# Pymaceuticals DataFrame Table:
print('\n\033[4mPymaceauticals Data Table\033[0m:\n') 
pymaceuticals_df
```

```python
# Duplicate Timepoints:
duplicate_timepoints = pymaceuticals_df[pymaceuticals_df.duplicated(subset = ['Mouse ID', 'Timepoint'], keep = False)]

# Duplicate Data Table:
print('\n\033[4mDuplicate Timepoints Data Table\033[0m:\n')
duplicate_timepoints
```

```python
# Pymaceuticals DataFrame (removed duplicate data from last step):
cleaned_pymaceuticals_df = pymaceuticals_df[~pymaceuticals_df['Mouse ID'].isin(duplicate_timepoints['Mouse ID'].unique())]

# Number of Unique Mouse IDs (after duplicates removed):
cleaned_unique_mice_id = cleaned_pymaceuticals_df['Mouse ID'].nunique()
print(f'\n\033[4mNumber of unique Mouse IDs\033[0m: {cleaned_unique_mice_id}')

# Cleaned Pymaceuticals DataFrame Table:
print('\n\033[4mCleaned Pymaceauticals Data Table\033[0m:\n')
cleaned_pymaceuticals_df
```

# Generate Summary Statistics
Create a DataFrame of summary statistics. Remember, there is more than one method to produce the results you're after, so the method you use is less important than the result.

1) Your summary statistics should include:
    - A row for each drug regimen. These regimen names should be contained in the index column.
    - A column for each of the following statistics: mean, median, variance, standard deviation, and SEM of the tumor volume.

```python
# Data:
grouped = cleaned_pymaceuticals_df.groupby ('Drug Regimen') ['Tumor Volume (mm3)']

# Calculations:
mean = grouped.mean()
median = grouped.median()
variance = grouped.var()
std_dev = grouped.std()
sem = grouped.sem()

# DataFrame:
summary_stats = pd.DataFrame ({
    'mean': mean,
    'median': median,
    'variance': variance,
    'std dev': std_dev,
    'SEM': sem
}).round(2)
summary_stats.index.name = 'Drug Regimen'
summary_stats.columns.name = 'Tumor Volume (mm3)'

# Summary Results:
print('\n\033[4mSummary Statistics\033[0m:\n')
summary_stats
```

# Calculate Quartiles, Find Outliers, and Create a Box Plot
1) Calculate the final tumor volume of each mouse across four of the most promising treatment regimens: Capomulin, Ramicane, Infubinol, and Ceftamin. Then, calculate the quartiles and IQR, and determine if there are any potential outliers across all four treatment regimens. Use the following substeps:
    - Create a grouped DataFrame that shows the last (greatest) time point for each mouse. Merge this grouped DataFrame with the original cleaned DataFrame.
    - Create a list that holds the treatment names as well as a second, empty list to hold the tumor volume data.
    - Loop through each drug in the treatment list, locating the rows in the merged DataFrame that correspond to each treatment. Append the resulting final tumor volumes for each drug to the empty list.
    - Determine outliers by using the upper and lower bounds, and then print the results.

2) Using Matplotlib, generate a box plot that shows the distribution of the final tumor volume for all the mice in each treatment group. Highlight any potential outliers in the plot by changing their color and style

hint: All four box plots should be within the same figure. Use this Matplotlib documentation pageLinks to an external site. for help with changing the style of the outliers.

```python
# Max Timepoint DataFrame:
last_timepoint_df = cleaned_pymaceuticals_df.groupby('Mouse ID')['Timepoint'].max().reset_index()
last_timepoint_df.columns = ['Mouse ID', 'Max Timepoint']
merged_pymaceuticals_df = pd.merge(cleaned_pymaceuticals_df, 
                                   last_timepoint_df, 
                                   how = 'right', 
                                   on = ['Mouse ID', 'Mouse ID']
                                  )
# Lists:
four_regimens_list = ['Capomulin', 
                      'Ramicane', 
                      'Infubinol', 
                      'Ceftamin'
                     ]
tumor_volume_data_list = []

# Filtered Regimens DataFrame:
four_regimens_df = merged_pymaceuticals_df[merged_pymaceuticals_df['Drug Regimen'].isin(four_regimens_list)]

# Filtered Regimens + Filtered Timepoint DataFrame:
filtered_final_timepoint = four_regimens_df[four_regimens_df['Timepoint'] == four_regimens_df['Max Timepoint']]

# For Loops:
for regimen in four_regimens_list:
    regimen_tumor_data = filtered_final_timepoint[filtered_final_timepoint['Drug Regimen'] == regimen]['Tumor Volume (mm3)']
    tumor_volume_data_list.append(regimen_tumor_data.tolist())
    
# Calculations:
def calculate_outliers(regimen_tumor_data):
    quartiles = regimen_tumor_data.quantile([0.25, 0.50, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)
    
    # Outliers:
    outliers = regimen_tumor_data[(regimen_tumor_data < lower_bound) | (regimen_tumor_data > upper_bound)]
    return lowerq, upperq, iqr, lower_bound, upper_bound, outliers

for regimen, volumes in zip(four_regimens_list, tumor_volume_data_list):
    volumes_series = pd.Series(volumes)
    lowerq, upperq, iqr, lower_bound, upper_bound, outliers = calculate_outliers(volumes_series)
    print(f'\n\033[4m{regimen} Summary\033[0m:\n') 
    print(f'Lower Quartile (Q1): {lowerq:.2f}')
    print(f'Upper Quartile (Q3): {upperq:.2f}')
    print(f'Interquartile Range (IQR): {iqr:.2f}')
    print(f'Lower Bound: {lower_bound:.2f}')
    print(f'Upper Bound: {upper_bound:.2f}')
    print(f'Outliers: {outliers}\n')
```

```python
# Box Plot:
plt.figure(figsize = (6,4))
plt.boxplot(tumor_volume_data_list, 
            labels = four_regimens_list, 
            patch_artist = True,
            flierprops = dict(marker = 'o', 
                            color = 'red', 
                            markersize = 6))
plt.title('\nFinal Tumor Volume by Drug Regimen\n')
plt.xlabel('\nDrug Regimen\n')
plt.ylabel('\nFinal Tumor Volume (mm3)\n')
plt.show()
```

# Calculate Correlation and Regression
1) Calculate the correlation coefficient and linear regression model between mouse weight and average observed tumor volume for the entire Capomulin treatment regimen.
2) Plot the linear regression model on top of the previous scatter plot.

```python
# Pearson Correlation Coefficient + Linear Regression Model:
correlation_coefficient = capomulin_merged['Weight (g)'].corr(capomulin_merged['Average Tumor Volume (mm3)'])
slope, intercept, r_value, p_value, std_err = st.linregress(
    capomulin_merged['Weight (g)'],
    capomulin_merged['Average Tumor Volume (mm3)']
)

# Regression Line:
regression_line = slope * capomulin_merged['Weight (g)'] + intercept

# Model Design:
plt.figure(figsize = (6,4))
plt.scatter(capomulin_merged['Weight (g)'],
            capomulin_merged['Average Tumor Volume (mm3)'], 
            color = 'black', 
            label = 'Data points')
plt.plot(capomulin_merged['Weight (g)'], 
         regression_line, 
         color = 'red', 
         label = 'Regression line')
plt.title('\nMouse Weight vs Average Tumor Volume for Capomulin Regimen\n')
plt.xlabel('\nWeight (g)\n')
plt.ylabel('\nAverage Tumor Volume (mm3)\n')
plt.show()

# Summary:
print('\033[4mSummary of Linear Regression\033[0m:')
print(f'\nCorrelation Coefficient: {correlation_coefficient:.2f}')
print(f'Slope: {slope:.2f}')
print(f'Intercept: {intercept:.2f}')
print(f'R-squared: {r_value**2:.2f}')
```
